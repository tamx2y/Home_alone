```python
# 분야 구분 함수
def 분야별지출(row, 분야):
    if 분야 == '장비':
        return sum([
            row['가계지출_소비지출_오락문화_영상음향기기구입비'],
            row['가계지출_소비지출_오락문화_사진광학장비구입비'],
            row['가계지출_소비지출_오락문화_정보처리장치구입비'],
            row['가계지출_소비지출_오락문화_기록매체구입비'],
            row['가계지출_소비지출_오락문화_영상음향정보기기수리비'],
            row['가계지출_소비지출_오락문화_악기기구구입비'],
            row['가계지출_소비지출_오락문화_내구재구입비'],
            row['가계지출_소비지출_오락문화_오락문화내구재유지수리비'],
            row['가계지출_소비지출_오락문화_캠핑운동관련용품구입비'],
            row['가계지출_소비지출_오락문화_장난감취미용품구입비']
        ])
    elif 분야 == '활동':
        return sum([
            row['가계지출_소비지출_오락문화_운동오락서비스이용금액'],
            row['가계지출_소비지출_오락문화_문화서비스이용금액'],
            row['가계지출_소비지출_오락문화_단체여행경비']
        ])
    elif 분야 == '동식물':
        return sum([
            row['가계지출_소비지출_오락문화_화훼관련용품구입비'],
            row['가계지출_소비지출_오락문화_애완동물관련물품구입비'],
            row['가계지출_소비지출_오락문화_화훼애완동물서비스이용금액']
        ])
    else:
        return 0

# apply로 변수 생성
young2['장비분야지출'] = young2.apply(lambda row: 분야별지출(row, '장비'), axis=1)
young2['활동분야지출'] = young2.apply(lambda row: 분야별지출(row, '활동'), axis=1)
young2['동식물분야지출'] = young2.apply(lambda row: 분야별지출(row, '동식물'), axis=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 중앙값 기준 연도별 취미지출 비율 계산 (퍼센트 변환)
grouped = young3.groupby('조사연도')[['소비지출총액', '취미비용', '실질취미지출']].median().reset_index()
grouped['명목취미비율'] = (grouped['취미비용'] / grouped['소비지출총액']) * 100
grouped['실질취미비율'] = (grouped['실질취미지출'] / grouped['소비지출총액']) * 100
grouped['명목기타'] = 100 - grouped['명목취미비율']
grouped['실질기타'] = 100 - grouped['실질취미비율']

# 2. 시각화
x = np.arange(len(grouped))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

# 명목 취미 지출 막대
bars1 = ax.bar(x - width/2, grouped['명목취미비율'], width, label='취미 지출 (명목)', color='salmon')
bars2 = ax.bar(x - width/2, grouped['명목기타'], width, bottom=grouped['명목취미비율'], label='기타 지출 (명목)', color='#FFE5E0')

# 실질 취미 지출 막대
bars3 = ax.bar(x + width/2, grouped['실질취미비율'], width, label='취미 지출 (실질)', color='lightblue')
bars4 = ax.bar(x + width/2, grouped['실질기타'], width, bottom=grouped['실질취미비율'], label='기타 지출 (실질)', color='#D6E9F8')

# 3. 텍스트 라벨 함수
def label_bars(bars, values, offset=1.5):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_y() + height / 2 + offset,
            f'{value:.1f}%',  # 퍼센트 표기
            ha='center', va='center', fontsize=10, color='black'
        )

label_bars(bars1, grouped['명목취미비율'])
label_bars(bars3, grouped['실질취미비율'])

# 4. 축 및 제목 설정
ax.set_ylabel('전체 소비지출 대비 취미지출 비율 (%)', fontsize=12)
ax.set_xlabel('조사연도', fontsize=12)
ax.set_title('중앙값 기준 연도별 취미비용 비율 (명목 vs 실질)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(grouped['조사연도'])
ax.set_ylim(0, 110)
ax.legend(loc='upper right')
ax.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = range(2019, 2025)
n_years = len(years)

fig, axes = plt.subplots(1, n_years, figsize=(4*n_years, 6), sharey=True)

for i, yr in enumerate(years):
    group_year = young3[young3['조사연도'] == yr]
    median_nominal = group_year['취미비용'].median()
    median_real = group_year['실질취미지출'].median()

    total = median_nominal + median_real
    # 100% 기준 비율 계산
    nominal_ratio = median_nominal / total
    real_ratio = median_real / total

    axes[i].bar(0, nominal_ratio, color='salmon', label='명목 취미비용')
    axes[i].bar(0, real_ratio, bottom=nominal_ratio, color='green', label='실질 취미비용')

    # 라벨링
    axes[i].text(0, nominal_ratio/2, f'{nominal_ratio:.1%}', ha='center', va='center', fontsize=10, color='black')
    axes[i].text(0, nominal_ratio + real_ratio/2, f'{real_ratio:.1%}', ha='center', va='center', fontsize=10, color='black')

    axes[i].set_title(f'{yr}년')
    axes[i].set_xticks([])
    if i == 0:
        axes[i].set_ylabel('비중 (100%)')

axes[0].legend(loc='upper right')
plt.suptitle('전체 청년 명목 및 실질 취미비용 비중 (100% 누적막대)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# 1. 중앙값 기준 연도별 취미 지출 비율 계산
grouped = young3.groupby('조사연도')[['소비지출총액', '취미비용']].median().reset_index()
grouped['명목취미비율'] = (grouped['취미비용'] / grouped['소비지출총액']) * 100
grouped['명목기타'] = 100 - grouped['명목취미비율']

# 2. 시각화
x = np.arange(len(grouped))
width = 0.5

fig, ax = plt.subplots(figsize=(10, 6))

# 누적 막대그래프: 취미비용(살몬), 소비지출 총액(회색)
bars1 = ax.bar(x, grouped['명목취미비율'], width, label='취미비용', color='salmon')
bars2 = ax.bar(x, grouped['명목기타'], width, bottom=grouped['명목취미비율'], label='총 소비지출 - 취미비용', color='lightgray')

# 취미비용 추세선 (검정색, 점선)
ax.plot(x, grouped['명목취미비율'], color='black', marker='o', linestyle='--', linewidth=1, label='취미비용 추세')

# 라벨링
for bar, val in zip(bars1, grouped['명목취미비율']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, f'{val:.1f}%', ha='center', va='center', fontsize=10, color='white')

for bar, val, bottom in zip(bars2, grouped['명목기타'], grouped['명목취미비율']):
    ax.text(bar.get_x() + bar.get_width() / 2, bottom + val / 2, f'{val:.1f}%', ha='center', va='center', fontsize=10, color='black')

# 축 및 제목 설정
ax.set_title('소비지출 총액 중 취미비용이 차지하는 비율', fontsize=14)
ax.set_xlabel('조사연도', fontsize=12)
ax.set_ylabel('비율 (%)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(grouped['조사연도'])
ax.set_ylim(0, 110)
ax.legend(loc='upper right')
ax.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
